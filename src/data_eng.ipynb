```python
import os
import json
from pprint import pprint
import pandas as pd
```

```python
base_data_path = '../data/'
total_jsons = len(os.listdir(os.path.join(base_data_path, 'jsons')))
print(f'Total JSON files: {total_jsons}')
```

```python
with open(os.path.join(base_data_path, 'jsons', '6.json'), 'rb') as f:
    tmp_json_data = json.load(f)
pprint(tmp_json_data) # doc_judgement
```

```python
### clean judgement text


def format_judgement(doc_judgement: list) -> str:
    formatted_text = ''.join(tmp_json_data['doc_judgement']).replace('\n', ' ').replace('\r', '')\
                    .replace('  ', ' ').replace('   ', '')
    return formatted_text


def add_clean_judgement_text(json_file_path: str) -> dict:
    with open(json_file_path, 'rb') as f:
        json_data = json.load(f)
    json_data['clean_judgement'] = format_judgement(json_data['doc_judgement'])
    return json_data

## exec

write_jsons_at = '../data/processed/p_jsons'

for json_file in os.listdir(os.path.join(base_data_path, 'jsons')):
    json_file_path = os.path.join(base_data_path, 'jsons', json_file)
    processed_json_data = add_clean_judgement_text(json_file_path)
    with open(os.path.join(write_jsons_at, json_file), 'w') as f:
        json.dump(processed_json_data, f)
    print(f'Processed and saved: {json_file}')
```
